```python
import numpy as np

A = np.array([[10,60,10,90],
              [20,50,40,70],
              [30,50,30,40],
              [20,50,20,60],
              [10,60,30,10]])
print("1-Computing Norm of each Instance:\n")

# 1-Computing Norm of each instance
i = 1;
for col in A.T:
  vec_norm = np.linalg.norm(col)
  print("Norm of Column " + str(i)+ " is: ",vec_norm)
  i +=1

print("\n3-Computing Euclidean Distance Matrix:\n")

# 3- Compute Euclidean Distance matrix of instances

M=N = A.shape[0];   #computing dimension of Distance Matrix
# Euclidean Distance Formula = (a-b)^2 = a.a + b.b - 2*(a.b)
#1- constructing a_dots matrix
a_dots = (A*A).sum(axis=1).reshape(M,1)*np.ones(shape=[1,N])
#constructing b_dots Matrix
b_dots = (A*A).sum(axis=1)*np.ones(shape=[M,1])
#Getting Final Euclidean Distance according to Formula
euclid_Dist = a_dots + b_dots - 2*(A.dot(A.T))
print(euclid_Dist)




  
```
